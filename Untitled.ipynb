```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv('sample_dataset_learnavi.csv', index_col=['Unnamed: 0'])
```

```python
tipe_source = []
tipe_target = []
```

```python
for a,b,c,d in zip(df.source.astype(str), df.source_name,df.target.astype(str), df.target_name):
    tipe_source.append(b[:-len(a)-1])
    tipe_target.append(d[:-len(c)-1])
```

```python
df['source_type'] = tipe_source
df['target_type'] = tipe_target
```

```python
s = df.to_dict('records')
```

```python
node = []
```

```python
for a in s:
    nodex = {'id': a['source_name'],'type':a['source_type']}
    node.append(nodex)
    nodex = {'id': a['target_name'],'type':a['target_type']}
    node.append(nodex)
```

```python
node = list({each['id'] : each for each in node}.values())
```
